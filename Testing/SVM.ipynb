```python
import numpy as np
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
```

```python
data = pd.read_csv("/Users/leoadlakha/Documents/Research Work/TDTVD/csv/data.csv")
```

```python
data = data.iloc[:, 2:]
```

```python
for i in data :
    print(data)
```

```python
y = data.iloc[:, 0]
y = np.array(y)
for i in range(len(y)) :
    if y[i] == 'Forged' :
        y[i] = 1 
    else :
        y[i] = 0 
Y = y.astype('int')
Y
```

```python
x = data.iloc[:, 1:]
x
```

```python
X = []
for i in range(len(data.iloc[:, 1:])) :
    X.append(np.array(data.iloc[i, 1:]))
```

```python
X
```

```python
svc = SVC(kernel='linear', C = 1)
svc.fit(X, Y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size = 0.8, shuffle = True)
```

```python
svc = SVC(kernel='linear', C = 0.001, gamma = 'auto')
svc.fit(x_train, y_train)
```

```python
svc.score(x_test, y_test)
```

```python
svc.score(x_train, y_train)
```

```python
clf = SVC()
grid = {'C' : [1e2, 1e3, 5e3, 1e4, 5e4, 1e5 ],
        'gamma' : [1e-3, 5e-4, 1e-4, 5e-3],
       }
abc = GridSearchCV(clf, grid, return_train_score=True, iid = True, cv = KFold(3, True, 0))
abc.fit(x_train, y_train)
```

```python
abc.best_estimator_
```

```python
abc.cv_results_
```
